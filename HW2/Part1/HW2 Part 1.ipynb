```python
import pandas as pd
## Replace path with your file's path
path = "../Part1/baywheels_ride_locations_2021.csv"
stations_imported = pd.read_csv(path, index_col=0, converters={"start_station_list": lambda x: x.strip("[]").replace("'","").replace("\"","").split(", ")})
stations_imported.info()
```

```python
s = stations_imported['start_station_list'].explode()
b = stations_imported[['start_date', 'start_hour']].join(pd.crosstab(s.index, s))
basket = b.drop(columns=['start_date', 'start_hour'])
for col in basket:
  basket[col] = basket[col].astype('bool')
```

```python
## Your code for LP goes here. 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(basket, min_support=0.28, use_colnames=True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 1) & (frequent_itemsets['support'] >= 0.28)]
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 2) & (frequent_itemsets['support'] >= 0.28)]
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 3) & (frequent_itemsets['support'] >= 0.28)]
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 4) & (frequent_itemsets['support'] >= 0.28)]
```

```python
f5 = frequent_itemsets[(frequent_itemsets['length'] == 5) & (frequent_itemsets['support'] >= 0.28)]
f5
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

rules
```
